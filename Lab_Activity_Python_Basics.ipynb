## 🧪 6. String Playground
Given the string `text = "Data Science is fun!"`, do the following:
- Convert to uppercase and lowercase
- Replace "fun" with "awesome"
- Slice the word "Science"
- Check if the word "Data" exists

```python
text = "Data Science is fun!";

upper_text = text.upper();
print(upper_text);

lower_text = text.lower();
print(lower_text);

replace_text = text.replace("fun", "awesome");
print(replace_text);

science_slice = text[5: 12]; # slice

print(science_slice);

print(text); # "Data" still exists in original
print(science_slice); # "Data" does not exist in slice
```

## 🧪 8. Built-In Fun
Use built-in functions to:
- Find the length of your full name
- Round the number 5.678 to 2 decimal places
- Get the maximum of the list `[12, 45, 23, 67, 34]`

```python
FLOAT_VALUE = 5.678;
LIST_OF_NUMS = [12, 45, 23, 67, 34];

def length_of_name(name):
  return len(name.strip('')); # trim whitespace

def round_to_nearest_hundredth(number):
  return(round(number, 2));

def maximum(list_of_numbers):
  return(max(list_of_numbers));


print(f"Count of letters in 'Rhea Mimi Carillo' (length): {length_of_name(FULL_NAME)}"); # see question 1 above

print(f"5.678 rounded to nearest hundredths: {round_to_nearest_hundredth(FLOAT_VALUE)}");

print(f"Max in list of [12, 45, 23, 67, 34]: {maximum(LIST_OF_NUMS)}");
```
